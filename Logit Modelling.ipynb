```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing Pre processed data

```python
df = pd.read_csv("Datapreprocessed.csv")
Original = df.copy()
#poutcome dropped because most of the fiels are unknown
df.drop("Poutcome", inplace = True , axis=1)
df.head()
```

```python
df.columns
```

```python
#shifting deposit to last position for convenience
columns = ['Age', 'Education', 'Default', 'Balance', 'Housing', 'Loan', 'Duration',
       'Campaign', 'Pdays', 'Previous', 
       'Job_management', 'Job_blue-collar', 'Job_technician', 'Job_admin',
       'Job_services', 'Job_retired', 'Job_self-employed', 'Job_student',
       'Job_umemployed', 'Job_entrepreneur', 'Job_housemaid', 'Married',
       'divorced', 'Contact_cellular', 'Contact_telephone', 'Season_1',
       'Season_2', 'First half of month', 'Deposit']
df = df.reindex(columns=columns)

df.head()
```

# Top features out of 29

```python

data  = df[['Education','Housing','Loan','Duration','Campaign', 'Previous','Job_retired','Job_student',
           'Married','Contact_cellular','Season_2','First half of month','Deposit']].copy()

#including all features in the model
X = data.iloc[:,0:12].values
Y = data.iloc[:,12:13].values
Y

```

```python
#logit code from statmodels library

import statsmodels.api as sm


logit = sm.Logit(endog=Y, exog = X)

result = logit.fit()

result.summary()

    
```

```python
#probit code from statmodels
probit = sm.Probit(endog=Y, exog = X)

result = probit.fit()

result.summary()
#result.pred_table(threshold=0.5)
```

```python
#using sklearn library

from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X, Y)

y_pred = classifier.predict(X)

# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y, y_pred)
cm
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y, classifier.predict(X))
```
